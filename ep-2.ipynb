```python
import  matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("WorldHappiness2018_Data.csv")
df
```

```python
rank = df['Rank']
score = df['Score']
```

```python
plt.plot(rank,score)
plt.title('Score Happy vs Rank')
plt.xlabel('Rank')
plt.ylabel('Score')
plt.show()
```

```python
gdp = df['GDP_Per_Capita']
lifeExp = df['Healthy_Life_Expectancy']

plt.plot(rank,gdp, color="green")
plt.plot(rank,lifeExp, color="red")
plt.title("World Country Score vs GDP and Life Exp")
plt.xlabel('Country Rank')
plt.legend()
plt.show()
```

```python
#grafik scatter plot

plt.scatter(gdp, score)
plt.title('GDP vs Happiness Score')
plt.xlabel('GDP')
plt.ylabel("Happines SCR")
plt.show()
```

```python
#Plot Histogram

plt.hist(score, color="maroon")
plt.title('Happiness Score Distribution')
plt.xlabel('GDP')
plt.ylabel("Happines SCR")
plt.show()
```

```python
#plot bar
#hindari membuat plot dari bool, float, string (x string, y stirng)

roundedScore = score.apply(int)
count = roundedScore.value_counts()
hapScore = count.index
```

```python
#plot bar
plt.bar(hapScore,count, color="maroon")
plt.title('Happiness Score')
plt.xlabel('Score')
plt.ylabel("Count")
plt.show()
```

```python
df2 = pd.read_csv('https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/retail_raw_reduced.csv')
df2
```

```python
df2.head()
```

```python
import datetime
df2['month'] = df2['order_date'].apply(lambda x : datetime.datetime.strptime(x, '%Y-%m-%d').strftime('%Y-%m'))
```

```python
df2['gmv'] = df2['item_price'] * df2['quantity']
df2
```

```python
#multiline Chart

df2.groupby(['month','brand'])['gmv'].sum().unstack().plot()
plt.title("Breakdown Brand by GMV 2019", loc="center", pad=20, fontsize=28, color='blue')
plt.xlabel('Order Month', fontsize=15)
plt.ylabel('Total Amount in Billion', fontsize=25)
plt.grid(color='darkgray', linestyle=":", linewidth=0.5)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1_000_000_000).astype(int))
plt.gcf().set_size_inches(10,5)
plt.tight_layout()
plt.legend(loc='right', bbox_to_anchor=(1.6,0.5), shadow=True, ncol=2)
```

```python
plt.clf()
df2.groupby(['month','province'])['gmv'].sum().unstack().plot(cmap='Set1')
plt.title("Breakdown Brand by GMV 2019", loc="center", pad=20, fontsize=28, color='blue')
plt.xlabel('Order Month', fontsize=15)
plt.ylabel('Total Amount in Billion', fontsize=25)
plt.grid(color='darkgray', linestyle=":", linewidth=0.5)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1_000_000_000).astype(int))
plt.legend(loc='lower center', bbox_to_anchor=(1.6,0.5), shadow=True, ncol=2, title='Province', fontsize=9, title_fontsize=18)
plt.gcf().set_size_inches(10,5)
plt.tight_layout()
```

```python
top_province = (df2.groupby('province'))['gmv'].sum().reset_index().sort_values(by='gmv', ascending=False).head(5)
top_province
```

```python
df2['province_top'] = df2['province'].apply(lambda x : x if (x in top_province['province'].to_list() ) else 'other' )
df2.groupby(['month', 'province_top'])['gmv'].sum().unstack().plot(marker='.', cmap='plasma')
plt.title("Breakdown Brand by Top Five Province 2019", loc="center", pad=20, fontsize=28, color='blue')
plt.xlabel('Order Month', fontsize=15)
plt.ylabel('Total Amount in Billion', fontsize=25)
plt.grid(color='darkgray', linestyle=":", linewidth=0.5)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1_000_000_000).astype(int))
plt.legend(loc='lower center', bbox_to_anchor=(1.6,0.5), shadow=True, ncol=2, title='Province', fontsize=9, title_fontsize=18)
plt.gcf().set_size_inches(10,5)
plt.tight_layout()
```

gcf = mengambil figure yang sudah diset sebelumnya<br>
clf = reset figure yand sudah diset

```python
#pie chart
#visualisasi data untuk mempresentasikan hasil perbandingan dari datasset bisa percentage, bisa variabel

df_dki_q4 = df2(df2['province'] == 'DKI Jakarta') & (df2['month']>='2019-10')
gmv_per_city_q4 = df_dki_q4.groupby('city')['gmv'].sum().reset_index()
plt.figure(figsize=(6,6))
plt.pie(gmv_per_city_q4['gmv'], label=gmv_per_city_q4['city'], autopct='%1.2f%%')
plt.title('GMV Contribution Per City - DKI Jakarta in 04-2019', loc='center', pad=30, fontsize=15, color='blue')
plt.show()
```
